```python
import spacy
import pandas as pd
import time
from docx import Document
# Load spaCy pipeline
nlp = spacy.load("es_core_news_lg")
```

```python
texto = ""
doc = Document(".\Libros\El Reino del Fuego.docx")
for paragraph in doc.paragraphs:
    texto += paragraph.text + "\n"
    
texto = texto.split("\n")
texto = list(filter(None, texto))
```

```python
libro = " ".join(texto)
libro.lower()
```

```python
t0 = time.time()
lemma_text_list = []
for doc in nlp.pipe(texto):
    lemma_text_list.append(" ".join(token.lemma_ for token in doc if token.is_alpha and not token.is_space and not token.is_stop and not token.is_punct))

t1 = time.time()
print("Total time: {}".format(t1-t0))
```

```python
" ".join(lemma_text_list)
```

```python



t0 = time.time()
frase = libro.lower()
doc = nlp(frase)
lista_limpia = [token.lemma_ for token in doc if token.is_alpha and not token.is_space and not token.is_stop and not token.is_punct]
df["original"] = pd.DataFrame({"original": lista_limpia})
t1 = time.time()
print("Total time: {}".format(t1-t0))  # ~9.759 seconds on 5000 rows


# Original method (very slow due to multiple calls to `nlp()`)
t0 = time.time()
df["text_lemma_1"] = df["text"].apply(lambda row: " ".join([w.lemma_ for w in nlp(row) if w.is_alpha and not w.is_space and not w.is_stop and not w.is_punct]))
t1 = time.time()
print("Total time: {}".format(t1-t0))  # ~9.759 seconds on 5000 rows


# Faster method processing rows as stream via `nlp.pipe()`
t0 = time.time()
lemma_text_list = []
for doc in nlp.pipe(df["text"]):
    lemma_text_list.append(" ".join(token.lemma_ for token in doc if token.is_alpha and not token.is_space and not token.is_stop and not token.is_punct))
df["text_lemma_2"] = lemma_text_list
t1 = time.time()
print("Total time: {}".format(t1-t0))
```
